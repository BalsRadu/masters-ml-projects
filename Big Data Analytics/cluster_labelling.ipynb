From the previous section, we know that the data in dataset 2 is not scaled; the values have different ranges, making the boxplot unclear. In the next section, we want to use clustering algorithms to find the best number of clusters, which might align with the number of target classes. However, these methods are distance-based (e.g., Euclidean distance), and the distance between points is influenced by the scale of the features. For example, features with larger ranges dominate the distance calculations. Thus, we need to scale the data before applying any clustering algorithm. While the Iris dataset does not strictly require scaling, we will scale it to maintain consistency.

For feature scaling, two common techniques are:

- **Normalization**: Scale features to [0, 1].
- **Standardization**: Z=(X)/ (mean = 0, variance = 1).

We will use the **Standardization** technique because it is less sensitive to outliers and clustering algorithms might perform better when features are in the same range. Besides, the features in dataset 2 are already normally distributed. For more details, click [here](https://www.geeksforgeeks.org/normalization-vs-standardization/).

To visualize potential outliers in dataset 2, we will standardize the data and plot the boxplot again.

# III. Clustering Analysis

We will use our own implementations of **KMeans** and **Fuzzy C-Means** algorithms to cluster the data and apply the **Elbow Method** to find the best number of clusters. The Elbow Method is a heuristic technique that uses the sum of squared distances between data points and their assigned clusters to determine the optimal number of clusters. The idea is to identify the "elbow point" where the sum of squared distances starts decreasing linearly. For more details, click [here](<https://en.wikipedia.org/wiki/Elbow_method_(clustering)>).

While we can visually inspect the Elbow plot, we will also use a quantitative approach. A naive method is calculating the derivative of the sum of squared distances to find the point with the steepest drop. Instead, we will use a dedicated package, **`kneed`**, designed to find the elbow point.


```python
from clustering.kmeans import KMeans
from clustering.fuzzy import FuzzyCMeans
from clustering.utils import calculate_sse_fuzzy, calculate_sse, find_elbow_point

def elbow_method(data: np.ndarray, max_k: int = 10, method: str = "kmeans"):
    """
    Applies the Elbow method to the given data, trying k=1..max_k 
    using the specified clustering method and plotting the SSE for each k.
    
    :param data: Dataset as (n_samples, n_features).
    :param max_k: Maximum number of clusters to try.
    :param method: Clustering method to use ("kmeans" or "fuzzy").
    """
    if method not in ["kmeans", "fuzzy"]:
        raise ValueError("Invalid method. Choose 'kmeans' or 'fuzzy'.")

    sse_values = []
    ks = range(1, max_k + 1)

    for k in ks:
        if method == "kmeans":
            kmeans = KMeans(n_clusters=k, max_iter=100, tolerance=1e-3)
            clusters, centers, labels = kmeans.run(data)
            sse = calculate_sse(clusters, centers, k)
        elif method == "fuzzy":
            fcm = FuzzyCMeans(n_clusters=k, max_iter=100, tolerance=1e-3, fuzzification=2.0)
            clusters, centers, labels, membership = fcm.run(data)
            sse = calculate_sse_fuzzy(data, centers, k, membership)
        sse_values.append(sse)

    # Find the optimal k using the Elbow method
    optimal_k = find_elbow_point(sse_values, ks, plot=True)
    print(f"Optimal number of clusters (k): {optimal_k}")

```

```python
elbow_method(iris_dataset, max_k=10, method="kmeans")
elbow_method(dataset2_np, max_k=10, method="kmeans")
```

```python
elbow_method(iris_dataset, max_k=10, method="fuzzy")
elbow_method(dataset2_np, max_k=10, method="fuzzy")
```

From the plots, the best number of clusters for the Iris dataset is 23 (depending on the algorithm used), consistent with the dataset's target classes. For dataset 2, the best number of clusters is consistently 3.

To validate our approach, we will use the best number of clusters from the Elbow Method to cluster the data using both KMeans and Fuzzy C-Means.


```python
from clustering.utils import plot_clusters

# Setting up parameters
k = 3

methods = {
    "kmeans": KMeans,
    "fuzzy": FuzzyCMeans
}

feature_names_iris=["Sepal length (cm)", "Sepal width (cm)"]

# Iterate through clustering methods
for method_name, ClusteringClass in methods.items():
    print(f"Processing using {method_name} clustering...")
    
    if method_name == "kmeans":
        clustering = ClusteringClass(n_clusters=k, max_iter=100, tolerance=1e-3)
        clusters, centers, labels = clustering.run(iris_np)
    elif method_name == "fuzzy":
        clustering = ClusteringClass(n_clusters=k, max_iter=100, tolerance=1e-3, fuzzification=2.0)
        clusters, centers, labels, membership = clustering.run(iris_np)

    # Plot the clusters
    plot_clusters(clusters, centers,  feature_names_iris, title=f"{method_name.capitalize()} Clustering on Iris Dataset")
```

# IV. Classifier Evaluation


While we have strong evidence that the clustering algorithms are working as expected and that we have identified the correct number of clusters, we still need to explore the practical value of using these algorithms to improve data labeling. Human experts are prone to errors and might mislabel data points, potentially causing learning models to falsely identify patterns in the data or fail to identify them altogether. This approach could also be valuable for large datasets where the sheer number of examples makes manual labeling impractical. However, we must remember that this might not be a universal solution, but we will explore it here to test our hypothesis and for educational purposes.

We will use the cleaned data (without outliers) to train the classifiers and evaluate their performance using the **accuracy score**, defined as the number of correct predictions divided by the total number of predictions. The data will be split into 80% for training and 20% for testing. Additionally, we will define a custom metric to compare the cluster labels with the true labels and count the number of correct predictions. This metric will be used to assess the performance of the clustering algorithms when labeling the data compared to the true labels from the dataset. We will also investigate how scaling the data affects the classifiers' performance.

To keep the code simple and ensure a fair comparison, we will leave most hyperparameters at their default settings. The resulting performance will serve as a baseline, which can be improved with hyperparameter tuning. At the end, we will rank the models based on their accuracy scores.

Our comprehensive analysis will include all combinations of the following:

- **Scalers**: None, StandardScaler, MinMaxScaler
- **Training algorithms**: RandomForest, KNeighbors, SVC, DecisionTree, LogisticRegression
- **Dimensionality reduction**: None, PCA

We will analyze the results, draw conclusions based on the accuracy score and the custom metric, and ultimately select the best model based on these metrics.


```python
# Load and preprocess the dataset
iris_df = pd.read_csv("datasets/Iris-150.csv", header=None)
iris_df.columns = ["Sepal length (cm)", "Sepal width (cm)", "Petal length (cm)", "Petal width (cm)", "Species"]
iris_df
```

```python
# Clean data
iris_df_cleaned = remove_outliers(iris_df.iloc[:, :-1])
iris_df_cleaned["Species"] = iris_df["Species"]

# Prepare features and labels
features = iris_df_cleaned.iloc[:, :-1]
labels = iris_df_cleaned["Species"]
labels_encoded = labels.astype("category").cat.codes

# Dataset to numpy array
iris_features_np = features.to_numpy()
iris_features_2d = reduce_to_2d(iris_features_np)

iris_features_np, iris_features_2d, labels_encoded
```

```python
from clustering.utils import compute_labelling_similarity, suppress_output


# Function to generate KMeans labels
def generate_kmeans_labels(features, true_labels, n_clusters=3, max_iter=100, tolerance=1e-3, similarity_threshold=0.75):
    """
    Generates KMeans labels with a similarity score to true labels above the given threshold.
    """
    while True:
        with suppress_output():
            # Perform KMeans clustering
            kmeans = KMeans(n_clusters=n_clusters, max_iter=max_iter, tolerance=tolerance)
            _, _, kmeans_labels = kmeans.run(features)

        # Compute similarity
        similarity = compute_labelling_similarity(kmeans_labels, true_labels)
        
        # Break if similarity threshold is met
        if similarity >= similarity_threshold:
            break

    return kmeans_labels, similarity
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Define scalers
scalers = {
    "None": None,
    "MinMaxScaler": MinMaxScaler(),
    "StandardScaler": StandardScaler(),
}

# Define classifiers
classifiers = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(random_state=42, probability=True),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
}

# Generate all feature-scaler-label combinations
combinations = []
for feature_name, feature_set in [("Normal Features", iris_features_np), ("2D Features", iris_features_2d)]:
    for scaler_name, scaler in scalers.items():
        # Apply scaling if a scaler is defined
        if scaler:
            scaled_features = scaler.fit_transform(feature_set)
        else:
            scaled_features = feature_set
        
        # Generate new KMeans labels for the scaled features
        kmeans_labels, similarity = generate_kmeans_labels(scaled_features, labels_encoded)
        
        # Add combinations
        combinations.append({
            "features": scaled_features,
            "labels": labels_encoded,
            "name": f"{feature_name} + {scaler_name} + Encoded Labels"
        })
        combinations.append({
            "features": scaled_features,
            "labels": kmeans_labels,
            "name": f"{feature_name} + {scaler_name} + KMeans Labels (Similarity: {similarity:.2f})"
        })

# Store results for each combination
all_results = {}

for combination in combinations:
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        combination["features"], combination["labels"], test_size=0.2, random_state=42
    )
    
    # Evaluate classifiers
    results = {}
    for name, clf in classifiers.items():
        # Train classifier
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        
        # Store accuracy
        results[name] = accuracy
    
    # Save results for the combination
    all_results[combination["name"]] = results

# Print results
for combination_name, metrics in all_results.items():
    print(f"\nResults for {combination_name}:")
    for model, accuracy in metrics.items():
        print(f"  {model}: Accuracy = {accuracy:.2f}")
```

### Results and Conclusions

The results of exploring the performance of the classifiers on the clustered data are as follows:

- There are no significant differences in the performance of the classifiers when using the Iris dataset, but **SVC** and **RandomForest** seem to perform slightly better than the other classifiers.
- The use of a scaler does not seem to affect the ability of the models to learn the target labels, but it does impact the custom metric. Scaling the data with either **StandardScaler** or **MinMaxScaler** decreases the number of correctly labeled points, with this effect being more pronounced when using PCA for dimensionality reduction.
- When using the labels generated by the clustering algorithms, the classifiers perform better than when using the true labels from the dataset. This compensates for the decrease in the number of correctly labeled points. In the best-case scenario, the performance of the classifiers is similar to that achieved using the true labels (0.93 vs 0.89). This is a strong indication that the clustering algorithms can effectively label data for a model.
- Using **PCA** for dimensionality reduction boosts the accuracy scores of the classifiers to nearly 1, despite the supposed loss of information during the compression process. By applying PCA, we can reduce the dimensions of the data and thus also lower the computational cost of the model.

In conclusion, the clustering algorithms can effectively label the data, and the performance of the classifiers using these labels is comparable to that achieved with the true dataset labels. The use of PCA for dimensionality reduction enhances the classifiers' performance, while the use of a scaler appears to diminish it. The best model combines **PCA** for dimensionality reduction, no scaler, and the **SVM** classifier, achieving an accuracy score of 1 and a custom metric of 0.89.


# V. Predicting the target classes


We will now move to the final part of the project: predicting the possible target classes for the second dataset. Using the 2D PCA-transformed data, we will train the best-performing classifier from the previous section and evaluate whether it can predict the classes identified by the clustering algorithms. Additionally, we will assess whether scaling the data for the second dataset affects the model's performance.


```python
# Load Dataset
ds2_df = pd.read_csv("datasets/Dataset2.csv", header=None)

# Clean data
ds2_cleaned = remove_outliers(ds2_df)

# Define scalers
norm = MinMaxScaler()
standard = StandardScaler()

# Dataset to numpy array
ds2_np = ds2_cleaned.to_numpy()
ds2_norm = norm.fit_transform(ds2_np)
ds2_standard = standard.fit_transform(ds2_np)

# Reduce to 2D
ds2_2d = reduce_to_2d(ds2_np)
ds2_2d_norm = reduce_to_2d(ds2_norm)
ds2_2d_standard = reduce_to_2d(ds2_standard)
```

```python
# Define a function to run KMeans and plot the results
def cluster_and_plot(data, title):
    """
    Perform KMeans clustering on the given data and plot the clusters.

    Parameters:
    - data: The input data for clustering.
    - title: Title of the plot.
    """
    # Clustering with KMeans
    kmeans = KMeans(n_clusters=3, max_iter=100, tolerance=1e-3)
    clusters_kmeans, centers_kmeans, labels_kmeans = kmeans.run(data)


    # Prepare clusters for plotting
    clusters_kmeans_plot = [
        [data[idx] for idx in range(len(labels_kmeans)) if labels_kmeans[idx] == c]
        for c in range(3)
    ]

    plot_clusters(clusters_kmeans_plot, centers_kmeans, feature_names_dataset2, title)

# Plot results for all datasets
cluster_and_plot(ds2_2d, "KMeans Clustering on Dataset 2 (2D)")
cluster_and_plot(ds2_2d_norm, "KMeans Clustering on Dataset 2 (Normalized 2D)")
cluster_and_plot(ds2_2d_standard, "KMeans Clustering on Dataset 2 (Standardized 2D)")
```

PCA expects the data to be scaled because it is sensitive to variance, and features with larger magnitudes can dominate the principal components. This did not affect the classifiers' performance in the previous section due to the small number of features and the relatively uniform range of the data. However, in this case, scaling might influence the model's ability to learn the true labels, depending on the label distribution. We will use **SVM** as the classifier, as it was the best-performing model in the previous section, along with PCA for dimensionality reduction and **StandardScaler** for scaling the data.


```python
from sklearn.metrics import classification_report

# Clustering with KMeans for `ds2_2d_standard`
kmeans = KMeans(n_clusters=3, max_iter=100, tolerance=1e-3)
_, _, labels_kmeans_standard = kmeans.run(ds2_2d_standard)

# Split the data
X_train_std, X_test_std, y_train_std, y_test_std = train_test_split(
    ds2_2d_standard, labels_kmeans_standard, test_size=0.2, random_state=42
)

# Clustering with KMeans for `ds2_np`
_, _, labels_kmeans_np = kmeans.run(ds2_np)

# Split the data
X_train_np, X_test_np, y_train_np, y_test_np = train_test_split(
    ds2_np, labels_kmeans_np, test_size=0.2, random_state=42
)

# Train and evaluate SVM
def train_and_evaluate_svm(X_train, X_test, y_train, y_test, title):
    """
    Trains an SVM classifier and evaluates its performance.
    Plots the decision boundary for 2D data.
    """
    # Train SVM
    svm = SVC(kernel='linear', random_state=42)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)

    # Compute metrics
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{title} - Accuracy: {accuracy:.2f}")
    print(classification_report(y_test, y_pred))

    # Visualize decision boundary (only for 2D data)
    if X_train.shape[1] == 2:
        h = 0.02  # Step size in the mesh
        x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
        y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
        Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)

        # Plot decision boundary
        plt.figure(figsize=(8, 6))
        plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
        plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolor='k', cmap=plt.cm.coolwarm)
        plt.title(f"SVM Decision Boundary ({title})")
        plt.xlabel("Feature 1")
        plt.ylabel("Feature 2")
        plt.show()


# Train and evaluate SVM for `ds2_np`
train_and_evaluate_svm(X_train_np, X_test_np, y_train_np, y_test_np, "Original Data")


# Train and evaluate SVM for `ds2_2d_standard`
train_and_evaluate_svm(X_train_std, X_test_std, y_train_std, y_test_std, "Standardized 2D Data")
```

As expected, the model's performance is very high in both cases, with a slight advantage when using the scaled 2D PCA data. As previously noted, the true accuracy is likely to be slightly lower than the result obtained here. However, this provides strong evidence that clustering algorithms can effectively label data for a model. Furthermore, we can conservatively conclude that the number of distinct classes is 3 (as indicated by the decision boundary and point distribution), and the model can predict these classes with high accuracy.

Overall, the preliminary results are very promising. They demonstrate the feasibility of creating a simple proof-of-concept, end-to-end pipeline to input a dataset, label it using clustering algorithms, and then train a model to predict those labels.
